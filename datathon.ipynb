```python
def extract_min_experience(value):
    import re
    # Find all numbers in the string
    numbers = re.findall(r'\d+', value)
    # Convert to integers and return the maximum if numbers are found
    if numbers:
        return min(map(int, numbers))
    # Default to 0 if no numbers are found
    return 0

# Apply the function to the column
data_new['start_dates'] = data_new['start_dates'].fillna(data_new['start_dates'].mode()[0]).apply(extract_min_experience)
data_new['passing_years'] = data_new['passing_years'].fillna(data_new['passing_years'].mode()[0]).apply(extract_min_experience)
data_new['can_exp']=(data_new['start_dates']-data_new['passing_years'])+1
data_new.drop(columns=['start_dates','passing_years'],inplace=True)
data_new['experiencere_requirement']=data_new['experiencere_requirement'].fillna(data_new['experiencere_requirement'].mode()[0]).apply(extract_min_experience)



# Display the updated DataFrame
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define the function to compute cosine similarity
def cosine_similarity_percentage(text1, text2):
    # Ensure both texts are strings (in case they are lists or other types)
    text1 = str(text1) if text1 is not None else ''
    text2 = str(text2) if text2 is not None else ''
    
    # Create TF-IDF vectors
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([text1, text2])
    
    # Compute cosine similarity
    similarity = cosine_similarity(vectors[0], vectors[1])
    return similarity[0][0] * 100  # Convert to percentage

# Assuming 'can_data' and 'required_data' are two DataFrames with columns 'skills' and 'combined_skills' respectively

# Create a new column 'similarity_score' to store the cosine similarity
can_data['similarity_score'] = [
    cosine_similarity_percentage(can_data['skills'].iloc[i], required_data['combined_skills'].iloc[i])
    for i in range(len(can_data))
]

# Print the resulting DataFrame
print(can_data[['skills', 'similarity_score']])
```
